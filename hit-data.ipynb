```python
# Cell 2: Imports
import pandas as pd
import numpy as np
from pybaseball import statcast_batter, playerid_lookup
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
```

```python
# Cell 3: Look up player IDs
def get_player_id(first_name, last_name):
    info = playerid_lookup(last_name, first_name)
    return int(info['key_mlbam'].values[0])

batter_id = get_player_id("Mike", "Trout")
pitcher_id = get_player_id("Gerrit", "Cole")
print("Batter ID:", batter_id)
print("Pitcher ID:", pitcher_id)
```

```python
# Cell 4: Get batter vs pitcher history
from pybaseball import cache
cache.enable()  # Cache data for faster re-runs

start_date = "2015-04-01"
end_date = "2024-07-01"

df = statcast_batter(start_date, end_date, batter_id)
df_vs_pitcher = df[df['pitcher'] == pitcher_id]

print(f"Matchups found: {len(df_vs_pitcher)}")
df_vs_pitcher[['game_date', 'events', 'description', 'pitch_type']].head()
```
